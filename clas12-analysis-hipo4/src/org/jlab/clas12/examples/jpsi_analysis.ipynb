# This is Figure 3 which graphs the z vertex distributions for electrons and positrons and shows the vertical cuts (-8cm<Vz<1cm)

```python
df['evz'].plot.hist(bins=200, alpha=0.5, range=(-15, 35.0), label='e-')
df['epvz'].plot.hist(bins=200, alpha=0.5, range=(-15, 35.0), label='e+')


plt.axvline(-8, color='r', linestyle='dashed', linewidth=2)
plt.axvline(1, color='r', linestyle='dashed', linewidth=2)
plt.xticks(np.arange(-15, 35, 5.0))


plt.legend()
plt.xlabel("z-vertex, cm")
plt.title("Fig. 3 Z-vertex distributions")
```

## This is Figure 4 which graphs the vertex time difference between the electron and the positron

```python
df['eepvtd'].plot.hist(bins=100, alpha=0.5, range=(-5, 5.0))
plt.yscale("log")

_ = plt.xlabel("t- - t+, ns")
_ = plt.title("Fig. 4 Vertex Time Difference")
```
